```python
import pandas as pd
from plotnine import ggplot, aes, geom_point, geom_smooth
```

# Sleep 

```python
sleep_health = pd.read_csv("sleep_health.csv")
```

```python
sleep_health.head()
```

```python
sleep_health.describe()
```

```python
sleep_health.shape
```

```python
sleep_health.isnull().sum()
```

```python
sleep_health.dtypes
```

```python
plot1 = (
    ggplot(sleep_health, aes(x='Age', y='Stress Level')) +
    geom_point() +
    geom_smooth(method='lm', formula='y ~ x', se=False) +
    geom_point(aes(y='Quality of Sleep'), shape='o', color='blue') 
)
plot1.show()
```

```python
plot1 = (
    ggplot(sleep_health, aes(x='Sleep Duration', y='Stress Level')) +
    geom_point() +
    geom_smooth(method='lm', formula='y ~ x', se=False) +
    geom_point(aes(y='Stress Level'), shape='o', color='blue') 
)
plot1.show()
```
